```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.imdb.com/list/ls091998340/?view=compact'
```

```python
headers = {'User-Agent': 'Mozilla/5.0'}

response = requests.get(url, headers=headers)
```

```python
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
def extract_movie(soup):
  data_list = []

  for data in soup.find_all('div', class_='sc-b189961a-0 iqHBGn cli-children'):
    title_data = data.find('h3', class_='ipc-title__text').text.split('.')
    title = title_data[1]
    position = title_data[0]
    metadata = data.find_all('span', class_='sc-b189961a-8 hCbzGp cli-title-metadata-item')
    year = metadata[0].text
    runtime = metadata[1].text
    try:
      rated = metadata[2].text
    except:
      rated = 'Not Rated'

    rating = data.find('span', class_='ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating').text.split()[0]

    result = [position,title,year,runtime,rated,rating]
    data_list.append(result)

  return data_list


def create_dataframe(data_list):
  df = pd.DataFrame(data_list, columns=['Position', 'Title', 'Year', 'Runtime', 'Rated', 'Rating'])
  return df

def save_to_csv(df, filename):
  df.to_csv(filename, index=False)

```

```python
def main():
  data_list = extract_movie(soup)
  df = create_dataframe(data_list)
  save_to_csv(df, 'imdb_movies_rating.csv')

  return df




```

```python
if __name__ == '__main__':
  main().head(10)
```
